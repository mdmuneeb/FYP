```python
import pandas as pd
import os
```

```python
DATA_DIR = "../xgb_dataset"
threshold = 0.6

splits = ["train", "val"]
```

```python
for split in splits:

    path = os.path.join(DATA_DIR, f"{split}_features.csv")

    df = pd.read_csv(path)
    print(f"\n==== {split.upper()} ====")
    print("Original:", df.shape)

    # confidence
    conf_mob = df[["mob_0","mob_1","mob_2"]].max(axis=1)
    conf_eff = df[["eff_0","eff_1","eff_2"]].max(axis=1)
    conf_res = df[["res_0","res_1","res_2"]].max(axis=1)

    mask = (
        (conf_mob < threshold) |
        (conf_eff < threshold) |
        (conf_res < threshold)
    )

    removed = mask.sum()

    df_clean = df[~mask].reset_index(drop=True)

    save_path = os.path.join(DATA_DIR, f"{split}_features_clean.csv")
    df_clean.to_csv(save_path, index=False)

    print(f"Removed: {removed}")
    print(f"Saved: {save_path}")
    print("New shape:", df_clean.shape)
```
